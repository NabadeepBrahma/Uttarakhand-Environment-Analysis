```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = r"C:\Users\nabad\Documents\Uttrakhand Environment Analysis\Uttarakhand_Environment_Analysis_Large.xlsx"
df = pd.read_excel(file_path)

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Define past (1980-2000) and present (2001-2025)
past = df[(df['Year'] >= 1980) & (df['Year'] <= 2000)]
present = df[(df['Year'] >= 2001) & (df['Year'] <= 2025)]

# Metrics to compare
metrics = ['Average Temperature (Â°C)', 'Rainfall (mm)', 'Humidity (%)', 'Forest Cover (sq km)', 
           'Water Quality Index', 'Groundwater Level (m)', 'Air Pollution Index', 'Waste Generation (tons)']

# Calculate statistical summaries
past_stats = past[metrics].agg(['mean', 'median', 'std'])
present_stats = present[metrics].agg(['mean', 'median', 'std'])

# Percentage change
percentage_change = ((present_stats - past_stats) / past_stats) * 100

# Visualization - Bar plots for percentage change
plt.figure(figsize=(12, 6))
percentage_change.loc['mean'].plot(kind='bar', color=['red' if x < 0 else 'green' for x in percentage_change.loc['mean']])
plt.axhline(0, color='black', linewidth=1)
plt.title("Percentage Change in Environmental Factors (Past vs. Present)")
plt.ylabel("Percentage Change (%)")
plt.xticks(rotation=45)
plt.show()

# Line plots for trends
plt.figure(figsize=(12, 6))
sns.lineplot(x=past['Year'], y=past['Average Temperature (Â°C)'], label="Past (1980-2000)", color='red')
sns.lineplot(x=present['Year'], y=present['Average Temperature (Â°C)'], label="Present (2001-2025)", color='blue')
plt.title("Temperature Trends: Past vs. Present")
plt.xlabel("Year")
plt.ylabel("Average Temperature (Â°C)")
plt.legend()
plt.show()
```
